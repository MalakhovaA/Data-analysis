```
new_df = new_df.drop('event_disp_name', axis=1)
new_df.sample(4)
```

Аналогично столбцы "event_name",	"primary_act_name" и "secondary_act_name". В таких случаях я буду оставлять тот столбец, в котором больше всего информации.

```
new_df = new_df.drop(['primary_act_name','secondary_act_name'], axis=1)
new_df.sample(4)
```

Объединю столбцы "venue_city" и "venue_state"

```
new_df['venue_city/state'] = new_df['venue_city'].str.cat(df['venue_state'], sep =" / ")
new_df = new_df.drop(['venue_city','venue_state'], axis=1)
new_df.sample(4)
```

Приводим колонки к корректным типам данных. 

```
new_df = new_df.drop(['event_dt'], axis=1)
```

```
new_df
```

Делим столбец sales_platform_cd на два: платформу и сайт.

```
new_df.sales_platform_cd.unique()
```

```
sales_platform_cd_sep = new_df['sales_platform_cd'].str.split('.',expand=True)
sales_platform_cd_sep.columns=['platform','site','a','b']
sales_platform_cd_sep.sample(6)
```

```
sales_platform_cd_sep['site'] = sales_platform_cd_sep['site'].str.cat(sales_platform_cd_sep['a'], sep =".").str.cat(sales_platform_cd_sep['b'],sep=".")
sales_platform_cd_sep = sales_platform_cd_sep.drop(['a','b'], axis=1)
sales_platform_cd_sep
```

```
new_df = new_df.join(sales_platform_cd_sep)
new_df
```

```
new_df = new_df.drop(['sales_platform_cd'], axis=1)
new_df.head()
```
